#### Part 1 data preparation

In the cell below, generate training data for your model.
<br> Generate a decent amount of training data in the interval of x.
<br>
<br>All data need to be separated by the same distance.


```python
import numpy as np

# Generate training data
x_train = np.linspace(-100, 100, 1000)[:, np.newaxis]  # Reshape to column vector
y_train = x_train * np.sin(x_train / 2300)
```

#### Part 2 build models

In the cells below, build your models and train it with data from part1.
<br>You need to split your training data to two parts. With 40% used in training, and 60% used in test.

<br>You may do this with sklearn's train_test_split() or keras' validation_split together with train_test_split().

```python

import tensorflow as tf
from sklearn.model_selection import train_test_split

# Split data
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.6)

# Model 1
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

model1.compile(optimizer='adam', loss='mse')
model1.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```

```python
# Model 2
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1)
])

model2.compile(optimizer='adam', loss='mse')
model2.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```

```python
# Model 3
model3 = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

model3.compile(optimizer='adam', loss='mse')
model3.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```

#### Part 3 model evaluation

In the cells below, eval your model with training data, test data ( with model.evaluate( ) ), and plot your predict result on the same plot with the plot of goal function.

```python
import matplotlib.pyplot as plt

# Evaluate model 1
loss1 = model1.evaluate(x_test, y_test)
print("Model 1 Test Loss:", loss1)

# Evaluate model 2
loss2 = model2.evaluate(x_test, y_test)
print("Model 2 Test Loss:", loss2)

# Evaluate model 3
loss3 = model3.evaluate(x_test, y_test)
print("Model 3 Test Loss:", loss3)

# Plot results
plt.scatter(x_test, y_test, label='Actual')
plt.scatter(x_test, model1.predict(x_test), label='Model 1')
plt.scatter(x_test, model2.predict(x_test), label='Model 2')
plt.scatter(x_test, model3.predict(x_test), label='Model 3')
plt.legend()
plt.show()



```

#### Part 4 get model output and feedforward by yourself

Recall how a Feedforward Neural Network gets its output. Now choose your model with highest accuracy, and call get_weights( ) to get its weights and bias.
<br> Hint: bias and weights have different dimensions in most cases. Try to guess which index represent bias before you look it up on the internet.

<br> Afterwards, choose 5 data from your training dataset, do all the necessary calculations (with program of course), and get the output of your model. Compare it to the model.predict( ) result.
<br> They should be identical for the first several digitals, if not for all digits.

```python
# Get weights and biases for the best model (say, model 1)
weights = model1.get_weights()

# Extract weights and biases
weights_hidden = weights[0]
biases_hidden = weights[1]
weights_output = weights[2]
biases_output = weights[3]

# Choose 5 data points for calculation
x_samples = x_train[:5]

# Manual feedforward calculation
hidden_layer_output = np.dot(x_samples, weights_hidden) + biases_hidden
hidden_layer_output = tf.nn.relu(hidden_layer_output)  # Apply ReLU activation function
output_manual = np.dot(hidden_layer_output, weights_output) + biases_output

# Model prediction
output_prediction = model1.predict(x_samples)

# Compare results
print("Manual Calculation Output:")
print(output_manual)
print("Model Prediction Output:")
print(output_prediction)
```
